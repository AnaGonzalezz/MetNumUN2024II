This consists of replacing missing values with other specific values. For example, the mean for numerical variables, or the most frequent value in the case of categorical variables.

````
import matplotlib.pyplot as plt

survived_counts = titanic['Survived'].value_counts()
labelsp=['Died ' + str(survived_counts[0]),'Survived ' + str(survived_counts[1])]
plt.figure(figsize=(5,3))
plt.pie(survived_counts,labels=labelsp,autopct='%1.1f%%')
plt.title('Died survived proportion',fontweight="bold")
plt.show()
````

**Analysis:**
The pie chart provides a stark visualization of the tragic outcome of the Titanic disaster, with a clear majority of passengers not surviving the incident. The data shows that out of the total passengers, a staggering 61.6% perished, represented by the large blue slice. In contrast, only 38.4% managed to survive, as depicted by the smaller orange segment. This dramatic disparity in survival rates underscores the enormity of the loss of life that occurred during the Titanic sinking.

````

age = titanic['Age']
plt.hist(age, 20, alpha=0.5, facecolor='#6495DF', rwidth=0.5)
plt.tight_layout()
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age distribution')
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.show()
````

**Analysis:** The age distribution graph shows that the majority of Titanic passengers were young adults. There's a big spike in the number of people between 20 and 30 years old, suggesting the ship was carrying a lot of young, adventurous travelers.
However, the graph also reveals an older group of passengers, with another noticeable peak in the 40 to 50 age range. This means the Titanic wasn't just carrying young thrill-seekers, but also had a significant number of older passengers, perhaps including families or business folks.
As you look at the higher age groups, the number of passengers drops off pretty quickly. There aren't many kids under 10 or elderly over 70 represented, hinting that these more vulnerable age groups may not have been well-represented on the doomed voyage.

````
from statsmodels.graphics.mosaicplot import mosaic
titanic["Died"] = titanic["Survived"]
titanic.loc[titanic['Survived'] == 0, 'Died'] = "Died"
titanic.loc[titanic['Survived'] == 1, 'Died'] = "Survived"
plt.rcParams['font.size'] = 10
plt.rcParams['figure.figsize'] = (5, 4)
mosaic(titanic.sort_values('Sex'), ['Sex', 'Died'])
plt.show()
````

**Analysis:** It presents a clear breakdown of the survival rates on the Titanic based on gender. The larger green rectangle indicates that the majority of male passengers died, while the smaller burgundy rectangle shows that a smaller proportion of female passengers perished. Conversely, the smaller burgundy rectangle for "female Survived" demonstrates that a higher percentage of women survived the disaster compared to men. This visual representation highlights the significant gender disparity in the Titanic's survival outcomes.

````

mosaic(titanic.sort_values('Pclass'), ['Pclass', 'Died'])
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (5, 4)
plt.show()
````

**Analyisis:** It provides a more detailed breakdown of the survival rates, categorizing the passengers by class. The red rectangles show the passengers who did not survive, with the largest rectangle representing those in the lowest passenger class. The green rectangles depict the survivors, with the largest rectangle indicating those in the highest passenger class who made it off the Titanic. This level of detail allows for a deeper understanding of how passenger class played a major role in the relative fates of those aboard the ill-fated ship.

````

mosaic(titanic.sort_values('Embarked'), ['Embarked', 'Died'])
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (5, 4)
plt.show()
````

**Analysis:** The largest teal rectangle indicates that a substantial number of passengers who boarded at Southampton (S) did not survive the disaster. In contrast, the smaller burgundy rectangles for Cherbourg (C) suggest fewer passengers from that port perished. The most promising outcome is seen for those who embarked at Queenstown (Q), with the green rectangle showing a higher proportion of survivors from that location. This disparity in survival rates across the different ports of origin highlights how the Titanic tragedy unfolded in a profoundly unequal manner, with passengers facing markedly different odds of making it off the ill-fated ship depending on where they joined the journey.
